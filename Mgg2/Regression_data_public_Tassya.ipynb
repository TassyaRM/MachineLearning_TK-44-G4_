```python
#Nama : Tassya Ramadhanti
#NIM  : 1103204016
#Regression DATA RUMAH

import pandas as pd

# Ganti path sesuai dengan lokasi dataset CSV di Google Drive Anda
file_path = '/content/DATA SAINS/DATA RUMAH (1).csv'

# Membaca dataset
df = pd.read_csv(file_path)
```

```python
#menampilkan 17 baris pertama dari DataFrame
df.head(17)
```

```python
#menampilkan data yang disimpan DataFrame
df
```

```python
#mendapatkan informasi dari DataFrame
df.info()
```

```python
#hapus kolom yang tidak penting untuk dipelajari model
data = df.drop(columns=['NO','NAMA RUMAH'])
```

```python
#one hot encoding dengan pd.get_dummies()
data = pd.get_dummies(data)
data
```

```python
# pisahkan atribut dan label
predictions = ['HARGA','LB','LT','KT','KM','GRS']
X = data[predictions]
y = data['GRS']
```

```python
# lakukan normalisasi terhadap data yang kita miliki
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X)
scaled_data = scaler.transform(X)
scaled_data = pd.DataFrame(scaled_data, columns = X.columns)
scaled_data.head()
```

```python
from sklearn.model_selection import train_test_split

#bagi data menjadi train dan test set
X_train, X_test, y_train, y_test = train_test_split(scaled_data, y, test_size=0.2, random_state=1)
```

```python
from sklearn import linear_model

# Membuat model logistic regression
model = linear_model.LogisticRegression()

# Latih model dengan fungsi fit()
model.fit(X_train, y_train)

# Mendapatkan parameter-parameter model
data = model.get_params()
print("LogisticRegression")
print(data)
```

```python
#uji akurasi model
model.score(X_test, y_test)
```
